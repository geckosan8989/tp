```python
import pandas as pd
```

```python
# Q.1
df = {"ProductID":[101,102,103,104,105],
     "Product":["Widget","Gadget","Doodad","Thingamajig","Whatchamacallit"],
     "Price": [10.99,24.95,5.99,15.50,8.75],
     "UnitsSold": [120,85,230,150,200]}
df1 = pd.DataFrame(data=df)
```

```python
# Q.2
print(df1)
```

```python
# Q.3
df1["Total_price"] = df1["Price"] * df1["UnitsSold"]
df1.head()
```

```python
#Q.4
avg_price = df1['Price'].mean()
print(avg_price)
```
